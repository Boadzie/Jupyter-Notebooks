## STEP 1 IMPORTING THE LIBRARIES

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
cancer = load_breast_cancer()
```

```
cancer.keys()
```

```
print(cancer["DESCR"])
```

```
print(cancer["target_names"])
```

```
print(cancer["feature_names"]) # the features
```

```
cancer["data"].shape
```

```
df_cancer = pd.DataFrame(np.c_[cancer["data"], cancer["target"]], columns=np.append(cancer["feature_names"], ["target"]))# learn more???
```

## STEP 3: VISUALIZING THE DATA

## STEP 4: MODEL TRAINING

```
X = df_cancer.drop(["target"], axis=1)
```

```
X.head()# x minus the target
```

```
y = df_cancer["target"]
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 5)# spliting the data into train and test sets
```

```
X_train
```

```
#Train model 
# importing the Support Vector Machine
from sklearn.svm import SVC
```

```
from sklearn.metrics import classification_report, confusion_matrix
```

```
svc_model = SVC()
```

```
svc_model.fit(X_train, y_train) # train the model
```

## STEP 5: MODEL EVALUATION

```
y_predict= svc_model.predict(X_test)
```

```
y_predict
```

```
cm = confusion_matrix(y_test, y_predict)
```

```
sns.heatmap(cm, annot= True)
```

## STEP 6: IMPROVING THE MODEL

```
min_train = X_train.min()
range_train = (X_train - min_train).max()
X_train_scaled = (X_train - min_train)/ range_train
```

```
sns.scatterplot(x=X_train["mean area"], y= X_train["mean smoothness"], hue="target", data= df_cancer)
```

```
sns.scatterplot(x=X_train_scaled["mean area"], y= X_train_scaled["mean smoothness"], hue="target", data= df_cancer)
```

```
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/ range_test
```

```
sns.scatterplot(x=X_test_scaled["mean area"], y= X_test_scaled["mean smoothness"], hue="target", data= df_cancer)
```

```
svc_model.fit(X_train_scaled, y_train)
```

```
y_predict = svc_model.predict(X_test_scaled)
```

```
cm = confusion_matrix(y_test, y_predict)
```

```
sns.heatmap(cm, annot=True)
```

```
print(classification_report(y_test, y_predict))
```
